```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io
```

```python
image_path = '/content/drive/MyDrive/ML AND DL/Datasets/IMG_20230825_080941.jpg'
image = io.imread(image_path)
```

```python
plt.figure(figsize = (8,4))
plt.subplot(1,2,1)
plt.title("orginal")
plt.imshow(image)
plt.axis('off')
```

```python
pixels = image.reshape((-1,3))

num_colors = 16
kmeans = KMeans(n_clusters=num_colors, random_state = 42)
kmeans.fit(pixels)

compressed_pixels = kmeans.cluster_centers_[kmeans.labels_].astype(int)
compressed_image = compressed_pixels.reshape(image.shape)
```

```python
plt.subplot(1,2,2)
plt.title(f"compressed image ({num_colors} colors)")
plt.imshow(compressed_image)
plt.axis('off')
plt.show()
```
