```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
data = pd.read_csv('/content/drive/MyDrive/ML AND DL/Datasets/zoo.csv')
print(data.info())
print(data.describe())
```

```python
data_features = data.drop(['animal_name'],axis=1)
```

```python
linkage_matrix = linkage(data_features,method='ward')
```

```python
plt.figure(figsize=(25,7))
dendrogram(linkage_matrix,labels=data['animal_name'].values,leaf_rotation=90,leaf_font_size=11)
plt.title('hierachical clustering dendrogram')
plt.show()
```
